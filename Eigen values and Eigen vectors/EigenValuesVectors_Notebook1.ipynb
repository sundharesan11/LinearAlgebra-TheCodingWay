Eigenvector and Eigenvalue

An eigenvector x is a non-zero vector that, when multiplied by the matrix A, results in a scaled version of itself. 
This relationship is represented as Ax = x, where  is the corresponding eigenvalue.

In a rectangular matrix, such as m  n where m  n, the operation Ax does not result in a scaled version of x because the dimensions are incompatible.

```python
import numpy as np

# Matrix A, vectors u and v.
A = np.array([[1, 6],[5, 2]])
u = np.array([6, -5])
v = np.array([3, -2])

# A*u and A*v
Au = np.dot(A, u)
Av = np.dot(A, v)

print(Av)
print(Au)
```

```python
import numpy as np

# function to find if a vector is a scalar multiple of another vector and find the scalar term
def find_scalar(vector1, vector2):
    if vector1.shape != vector2.shape:
        raise ValueError("Vectors must have the same dimensions.")
    
    scalar_factor = None
    
    for comp1, comp2 in zip(vector1, vector2):
        if comp2 == 0:
            return None # to aviod zero division
        if scalar_factor is None:
            scalar_factor = comp1 / comp2
        elif comp1 / comp2 != scalar_factor:
            return None
    
    return scalar_factor
```

```python
factor = find_scalar(Au, u)
if factor is not None:
    print("Au is a scalar multiple of u with a factor of:", factor)
    y1 = factor
else:
    print("Au is not a scalar multiple of u.")
```

```python
factor = find_scalar(Av, v)
if factor is not None:
    print("Av is a scalar multiple of v with a factor of:", factor)
    y2 = factor
else:
    print("Av is not a scalar multiple of v.")
```

Hence v is not a eigenvector of given matrix A.

```python
# function to identify if a vector is eigenvector or not given the matrix, vector and the eigenvalue(scalar)
def is_eigenvector(A,u,y):
    return np.array_equal(np.dot(A, u), y*u)
```

```python
is_eigenvector(A,u,y1)
```

Hence u is a eigenvector of given matrix A.

Proving the findings visually

```python
# Plotting the vectors and the corresponding labels with appropriate colors
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot([0, u[0]], [0, u[1]], 'b', label='u')
plt.plot([0, v[0]], [0, v[1]], 'r', label='v')
plt.plot([0, Au[0]], [0, Au[1]], 'b--', label='A*u')
plt.plot([0, Av[0]], [0, Av[1]], 'r--', label='A*v')
plt.text(u[0], u[1], 'u', fontsize=12, color='b', ha='right', va='bottom')
plt.text(v[0], v[1], 'v', fontsize=12, color='r', ha='right', va='bottom')
plt.text(Au[0], Au[1], 'A*u', fontsize=12, color='b', ha='right', va='bottom')
plt.text(Av[0], Av[1], 'A*v', fontsize=12, color='r', ha='right', va='bottom')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Vectors u, v, Au, Av')
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()
```

The vectors u (in blue) and v (in red), as well as their transformations (Au and Av) under A (dashed lines) are plotted. If the dashed lines are parallel to the original vectors and only scaled by a factor, then u and v are eigenvectors of A.

Clearly Au(blue dashed line) is a scaled vector of u but not the Av(red dashed line) of v vector.

Result:
    
    u is a eigenvector of A with the corresponding eigenvalue -4.
    v is not a eigenvector of A
