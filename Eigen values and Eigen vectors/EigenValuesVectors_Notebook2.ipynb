step 1: Import the required libraries

```python
import numpy as np
import sympy as sp
```

step 2: Define the matrix in  the statement above

```python
A = np.array([[2, 1, 0],
              [1, 3, -1],
              [0, -1, 4]])
```

step 3: The eigen values and the eigen vectors can be obtained using the numpy function linalg.eig()

        In NumPy we can compute the eigenvalues and right eigenvectors of a given square array with the help of numpy.linalg.eig(). It will take a square array as a parameter and it will return two values first one is eigenvalues of the array and second is the right eigenvectors of a given square array.

```python
eigenvalues, eigenvectors = np.linalg.eig(A)
```

```python
# Printing the eigenvalues
eigenvalues
```

```python
#Printing the eigenvectors
eigenvectors
```

####  Check if the eigenvectors corresponding to distinct eigenvalues are linearly independent.

step 4: With the help of sympy.Matrix().rref() method, we can put a matrix into reduced Row echelon form. Matrix().rref() returns a tuple of two elements. The first is the reduced row echelon form, and the second is a tuple of indices of the pivot columns.

```python
rref = sp.Matrix(eigenvectors).rref()
```

```python
rref[0]
```

```python
rref[1]
```

```python
rank = sp.Matrix(eigenvectors).rank()
```

```python
rank
```

    i) In reduced row echelon form, if there are rows of zeros, it implies that those rows correspond to redundant equations, which are linear combinations of the other equations. This redundancy indicates linear dependence.
    
    ii) Conversely, if there are no rows of zeros in the reduced row echelon form, each equation is independent, and the system is linearly independent.

    So, in summary, when analyzing the linear independence of a system of equations, one can examine the reduced row echelon form of the corresponding matrix. If there are no rows of zeros, the system is linearly independent; if there are rows of zeros, the system is linearly dependent.

The rref of the eigenvectors matrix has no rows of zeros, thus the system is linearly independent.
