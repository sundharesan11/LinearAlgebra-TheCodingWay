Formula to find the inverse of the matrix:

|A|    -> Determinant of matrix, A.

Adj(A) -> Adjoint of a matrix, A.

```python
# Python program to find determinant of a matrix with dimension 2x2.
def det2(m):
    return (m[0][0]*m[1][1] - m[1][0]*m[0][1])
```

```python
# Python program to find determinant of a matrix with dimension 3x3
```

```python
def det3(m):
    a = m[0][0]
    b = m[0][1]
    c = m[0][2]
    d = m[1][0]
    e = m[1][1]
    f = m[1][2]
    g = m[2][0]
    h = m[2][1]
    i = m[2][2]
    return (a*e*i) - (a*f*h) - (b*d*i) + (b*f*g) + (c*d*h) - (c*e*g)
```

```python
# Python program to find transpose of the matrix

def transpose(m):
    r = len(m)
    c = len(m[0])
    tm = []
    for i in range(c):
        row = []
        for j in range(r):
            row.append(m[j][i])
        tm.append(row)
    return tm  
```

#### Python program to find the product of two matrices( matrix A and its inverse)

```python
def matrix_multiplication(m1, m2):
    result = []
    if len(m1[0])!=len(m2):
        return "Matrix multiplication cannot be done with this two matrices"
    for i in range(len(m1)):
        row = []
        for j in range(len(m2[0])):
            res = 0
            for k in range(len(m1)):
                res += m1[i][k] * m2[k][j]
            row.append(res)
        result.append(row)
    return result
```
