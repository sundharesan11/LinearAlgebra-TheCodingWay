### Given: 
    A Probability matrix where each column represents the probablity of landing on a certain page in a row given from a 
    certain page that is present column.
    A vector x, representing probability of landing on a certain page.
    4 pages in the network.
    
### Tasks:
        i) Start from page1 move to the next page. Check the compositon of vector x after 100 iterations.
        i) Start from page3 move to the next page. Check the compositon of vector x after 100 iterations.

```python
P = [[0,        0,   1, 0.5],
     [0.333333, 0,   0, 0  ],
     [0.333333, 0.5, 0, 0.5],
     [0.333333, 0.5, 0, 0  ]]
```

Having a probability matrix P, and a vector x, representing probability of landing on certain page, we can compute the probability of landing on each page in the network by multiplying this vector and the multiplication matrix.

Defining a function to do matrix and vector multiplication.

```python
def matrix_vector_multiplication(m, v):
    result = []
    if len(m) != len(v):
        return "Matrix vector multiplication is not possible!"
    for i in range(len(m)):
        r = 0
        for j in range(len(m[0])):
            r += m[i][j] * v[j]
        result.append(r)
    return result
```

Task 1:
    We start from page 1 and hence the vector representing the probabilities is [1, 0, 0, 0] as we are certain in first page among the 4 pages in the network.

```python
x1 = [1, 0, 0, 0]
```

```python
x1 = matrix_vector_multiplication(P, x1)
x1
```

Hence from page 1, the probabilities of moving to other pages in the network is [0.0, 0.333333, 0.333333, 0.333333].

Repeating the above process 100 times.

```python
for i in range(100):
    x1 = matrix_vector_multiplication(P, x1)
```

```python
x1
```

Task 2:
    We start from page 3 and hence the vector representing the probabilities is [0, 0, 1, 0] as we are certain in third page among the 4 pages in the network.

```python
x2 = [0, 0, 1, 0]
```

```python
x2 = matrix_vector_multiplication(P, x2)
x2
```

Hence from page 3, the probabilities of moving to other pages in the network is [1.0, 0.0, 0.0, 0.0] as third page has only one outllink pointing to the first page.

Repeating the above process 100 times.

```python
for i in range(100):
    x2 = matrix_vector_multiplication(P, x2)
```

```python
x2
```

Here both the vectors x1 and x2 have almost same values indicating that after 100 iterations the probabilities of moving to next page is same when starting from page1 and page3.

This steady probability vector is called **steady-state vector** of the network which remains the same after several iterations in the network.
